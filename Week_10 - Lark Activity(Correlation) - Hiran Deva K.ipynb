# 4. Correlation Analysis

```python
import scipy 
from scipy.stats.stats import pearsonr
```

### a) Pearson correlation coefficient between ph and Chloramines

```python
pr1, p1 = pearsonr(data['ph'], data['Chloramines'])
print("Pearson's R coefficient between ph and Chloramines: ",pr1.round(3))
```

The Pearson's R coefficient between ph and Chloramines is close to zero or approximately zero. Hence ther's no linear correlation exist between ph and Chloramine.

### b) Pearson correlation coefficient between ph and Conductivity

```python
pr2, p2 = pearsonr(data['ph'], data['Conductivity'])
print("Pearson's R coefficient between ph and Conductivity: ",pr2.round(3))
```

The Pearson's R coefficient between ph and Conductivity is also close to zero or approximately zero. Hence ther's no linear correlation exist between ph and Conductivity.
