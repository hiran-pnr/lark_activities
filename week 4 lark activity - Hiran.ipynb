```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Telco = pd.read_csv("Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv")
Telco.head()
```

```python
Telco.info()
```

# 1. Compare churn count with respect to gender

```python
plt.figure(figsize=(10,6))
sns.countplot(x=Telco['gender'], hue=Telco['Churn'])
plt.title('Churn count vs Gender', fontweight='bold', fontsize=18)
plt.grid()
plt.show()
```

# 2. Find out how many female senior citizens there in the dataset

```python
fsc = Telco[(Telco['gender']=='Female') & (Telco['SeniorCitizen']==1)]
fsc
```

```python
plt.figure(figsize=(3,6))
sns.countplot(x=fsc['gender'], color='green')
plt.title('count of female senior citizens', fontweight='bold', fontsize=15)
plt.show()
```

# 3. Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(10,6))
plt.scatter(Telco['tenure'], Telco['TotalCharges'], s=8)
plt.title('Tenure vs Total Charges', fontweight='bold', fontsize=15)
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
plt.show()
```

# 4. Find out which contract is preferred by the senior citizen

```python
sc= Telco[Telco['SeniorCitizen']==1]
sc
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=sc['Contract'])
plt.title('Contract preference of Senior Citizen', fontweight='bold', fontsize=18)
plt.grid()
plt.show()
```

# 5. Comment on your finds on Payment Method

```python
plt.figure(figsize=(10,8))
sns.countplot(x=Telco['PaymentMethod'])
plt.title('Prefered payment method', fontweight='bold', fontsize=18)
plt.xticks(rotation=5)
plt.grid()
plt.show()
```
